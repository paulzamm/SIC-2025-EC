## Instrucciones

1. Importe los módulos que te permitan trabajar con fechas y tiempo (modulo datetime).

2. Use los diferentes métodos relacinados a operaciones con fechas y hora.

3. En el siguiente apartado revisa cada una las fechas contenidas en la lista e imprima el número de semanas y los dias restantes para llegar a esa fecha.

```python
# Espacio disponible para desarrollar el código y agregar los comentarios necesarios.
from datetime import datetime

lista_de_fechas = ['04/01/2024', '01/28/2024', '03/16/2024', '05/24/2024', '07/30/2025']
lista_de_fechas_ordenadas = sorted(lista_de_fechas, key=lambda fecha: datetime.strptime(fecha, '%m/%d/%Y'))
print("Fechas ordenadas:", lista_de_fechas_ordenadas)
print("-"*100)

def diff_fechas(lista):
    for i in range(len(lista) - 1):
        fecha_inicio = datetime.strptime(lista[i], '%m/%d/%Y')
        fecha_fin = datetime.strptime(lista[i + 1], '%m/%d/%Y')
        diff = fecha_fin - fecha_inicio
        semanas = diff.days // 7
        dias = diff.days % 7
        
        print(f"Fecha Inicio: { fecha_inicio.date().strftime('%m/%d/%Y') }, Fecha Fin: { fecha_fin.date().strftime('%m/%d/%Y') }, Semanas de diferencia: { semanas } con { dias } días")

diff_fechas(lista_de_fechas_ordenadas)
```

# Actividad: Manejo de la biblioteca Pandas.


## Instrucciones


1. Crear una serie de pandas, donde su indice este compuesto por fechas y en columna principal almace los valoras cargados en una lista.

2. Graficar los valores de la lista en función de la fecha.


```python
# Espacio para desarrollar el código y agregar los comentarios necesarios.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

precios = [140, 130, 120, 115, 160, 110, 170, 150, 157, 135]
# Estableciendo los rangos de las fechas
rango_fechas = pd.date_range(start='01/01/2025', end='31/12/2025')

# Seleccionar las fechas para los precios
fechas = np.random.choice(rango_fechas, size=len(precios), replace=False)

# Creando la serie con fechas como índice, y de manera ordenada
serie = pd.Series(precios, fechas).sort_index()

print("Serie de precios con índices de fechas:")
print(serie)

plt.figure(figsize=(10, 5))

# Gráfico de líneas
serie.plot(
    marker='o', 
    linestyle='-',
    linewidth='2',
    color='teal'
)
plt.title("Precios en función de las fechas")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

# Actividad: Operaciones con Data Frames


## Instrucciones


1. Dado el siguiente diccionario construya un objeto DataFrame.
2. Filtre los resultados por año de publicación.
3. Obtenga algunas columnas basado en que contenga alguna palabra en el nombre de la publicacion.


```python
import pandas as pd

publicaciones_literarias = {
    1605: "Don Quijote de la Mancha - Miguel de Cervantes",
    1865: "Alicia en el país de las maravillas - Lewis Carroll",
    1949: "1984 - George Orwell",
    1954: "El Señor de los Anillos - J.R.R. Tolkien",
    1997: "Harry Potter y la piedra filosofal - J.K. Rowling",
    2003: "El Código Da Vinci - Dan Brown"
}


# Espacio para desarrollar el código y agregar los comentarios necesarios.

# 1. Construyendo un DataFrame a partir de un diccionario
df = pd.DataFrame(publicaciones_literarias.items(), columns=["Año", "Nombre_publicación"])
print("Creando DataFrame a partir de un diccionario:")
print(df)

# 2. Filtrar resultados por año de publicación
filtrado_por_anio = df[df['Año'] > 1900]
print("\nFiltrando publicaciones posteriores a 1900:")
print(filtrado_por_anio)

# 3. Obtener columnas que contengan alguna palabra en el nombre de la publicacións
filtrado_por_columna = df[df['Nombre_publicación'].str.contains("vinci", case=False)]
print("\nFiltrando por palabra clave (vinci) en la publicación:")
print(filtrado_por_columna)
```
