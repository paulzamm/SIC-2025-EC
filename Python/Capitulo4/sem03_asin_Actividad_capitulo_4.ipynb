# Actividad: Implementación de una Pila en Python

En esta actividad, tendrás la oportunidad de aplicar tus conocimientos sobre estructuras de datos y programación en Python para implementar una pila (stack). Una pila es una estructura de datos que sigue el principio "Last-In-First-Out" (LIFO), lo que significa que el último elemento que se agrega es el primero en ser eliminado.

## Resumen:

- **stack** = [ ]  # Recordando que una pila es básicamente una lista.
- **append**: Método utilizado para agregar un elemento a la lista.
- **pop**: Método usado para eliminar elementos de la lista, sino se especifica el índice se elimina el último elemento y retorna la lista resultante.

```python
nombres_aleatorios = ["Ana", "Carlos", "David", "Elena", "Fernando", "Gabriela",
                      "Hugo", "Isabel", "Juan", "Luis", "Lucía",
                      "Lorena", "Manuel", "María", "Miguel", "Natalia",
                      "Nicolás", "Nora", "Óscar", "Olivia", "Pablo", "Paula",
                      "Pedro", "Raquel", "Rafael", "Renata", "Roberto", "Rosario",
                      "Sara", "Sergio", "Silvia", "Tomás", "Teresa", "Víctor", "Verónica",
                      "Ximena", "Xavier", "Yolanda", "Yenny", "Zaira", "Zoe", "Zacarías"]

# Generar un código para revisar los nombres contenidos en la lista, y guardarlos en diferentes listas,
# de acuerdo a la inicial del nombre, use los comandos usandos anteriormente y borra los elementos usando
# el comando "for nombre in nombres_aleatorios:" al final del proceso la lista inicial debe quedar vacia.
listas_por_inicial = {}

def nombres_pila(pila, listas):
  """Extraer nombres de una lista y guardarlas en diferentes listas de acuerdo a su inicial"""
  # Caso base, si la pila esá vacía termina la recursión
  if not pila:
    return

  nombre = pila.pop()  # Eliminar y devolver el (último) elemento tope de la pila
  inicial = nombre[0]  # Extraer la inicial del nombre

  if inicial not in listas:
    listas[inicial] = []  # Insertar las iniciales al dict

  listas[inicial].append(nombre)  # Insertar nombre en su respectiva inicial

  nombres_pila(pila, listas)  # Llamada recursiva


nombres_pila(nombres_aleatorios, listas_por_inicial)


for inicial, lista in listas_por_inicial.items():
  print(f"{inicial}: {', '.join(lista)}")

# Verificar que los elementos de la lista se hayan eliminado
print("\nLista inicial: ", nombres_aleatorios)
```

En este codódigo se muestra la creación de una clase para crear una pila y hacer las actividades básicas correspondientes.

- **__init__**: Creación de la pila.
- **is_empty**: Verificación si la pila está vacia.
- **push**: agregar elemento a la pila
- **pop**:si la pila no está vacia eliminar el último de elemento.

Use la notación mostrada en la presentacion 40, realice varias operaciones y comente brevemente los resultados (máximo 3 lineas), antes de pasar a la siguiente actividad asegurese de entender completamente la notación utilizada.

```python
# Escriba su código en está celda y todos los comentarios necesarios en la misma celda.

class Stack:
  def __init__(self):
    self.stack = []

  def is_empty(self):
    """ Devuelve true si la pila está vacía, caso contrario devuelve Falso """
    return True if len(self.stack) == 0 else False

  def push(self, item):
    """ Agrega un elemento al tope de la pila """
    self.stack.append(item)

  def pop(self):
    """ Elimina y devuelve el topa de la pila """
    return None if self.is_empty() else self.stack.pop()


stack = Stack()  # Creando la pila
stack.push("A")  # Insertando un ítem a la pila
stack.push("B")  # Insertando un ítem a la pila
print("Elemento tope: ", stack.pop())  # Eliminando y extrayendo el tope de la pila
stack.push("C")
print("Elemento tope: ", stack.pop())
print("Elemento tope: ", stack.pop())
print("Elemento tope: ", stack.pop())
print("La pila está vacía? ", stack.is_empty()) # Verificar si la pila está vacía
```

# Actividad: Búsqueda secuencial.

Se tiene una lista de 10 ciudades organizadas en función de la densidad de población, escriba una función para retornar la posición ocupada.


```python
ciudades = [
    "Daca, Bangladés",
    "Karachi, Pakistán",
    "Nueva Delhi, India",
    "Manila, Filipinas",
    "Seúl, Corea del Sur",
    "Cantón, China",
    "Taipéi, Taiwán",
    "Chenaral, Chile",
    "Shenzhen, China",
    "Bombay, India"
]

def obtener_posicion(ciudades, ciudad):
  """Obtener la posición ocupada de un elemento en una lista"""
  posicion = -1
  for i, valor in enumerate(ciudades):  # Iterando índice y valor de la lista
    if valor == ciudad:    # Comparar el elemento iterado con el elemento a buscar
      posicion = i  # Si encuentra el elemento devuelve la posición
  return posicion   # Si no encuentra devuelve (-1)

posicion_ocupada = obtener_posicion(ciudades,"Seúl, Corea del Sur")
print("Posición del elemento buscado: ", posicion_ocupada)
```

```python
# Espacio para replicar el código.

def do_experiment(floor, breaking):
  return floor >= breaking

def find_highest_safe_floor2(height, breaking):
  low, high = 1, height
  while low < high:
    mid = (low + high) // 2
    if do_experiment(mid, breaking):
      high = mid
    else:
      low = mid + 1
  return low


# Espacio para explicar funcionamiento:
# El código usa búsqueda binaria para encontrar el primer piso donde el huevo se rompe:
# divide el rango de pisos a la mitad en cada paso, verificando con (do_experiment).
# El piso seguro será low - 1.

# Cálculo y verificación del número de veces que se realiza la verificación.
def do_experiment(floor, breaking):
    return floor >= breaking

def find_highest_safe_floor2_checks(height, breaking):
    low, high = 1, height
    checks = 0
    while low < high:
        mid = (low + high) // 2
        checks += 1
        if do_experiment(mid, breaking):
            high = mid
        else:
            low = mid + 1
    return low - 1, checks

# Ejemplo
height = 100
breaking = 93
safe_floor, revisiones = find_highest_safe_floor2_checks(height, breaking)
print(f"El huevo estará a salvo hasta el piso {safe_floor}, Nro. Revisiones = {revisiones}")
```
