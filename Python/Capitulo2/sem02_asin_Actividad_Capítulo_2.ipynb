# Actividad: Creación de una Base de Datos en Python

## Descripción

A continuación debes crear una base de datos simple utilizando listas en Python. Una base de datos es una colección de datos organizados que se utilizan para almacenar y recuperar información. En este caso, simularemos una base de datos de estudiantes con información básica.

## Instrucciones

1. Crea un programa en Python que permita a los usuarios realizar las siguientes acciones:

   a. Agregar un nuevo estudiante a la base de datos. Cada estudiante debe tener al menos un nombre, un número de identificación y una edad. Puedes utilizar una lista para almacenar estos datos.

   b. Mostrar la lista de estudiantes en la base de datos.

   c. Buscar un estudiante por su número de identificación y mostrar su información.

   d. Actualizar la información de un estudiante existente en la base de datos.

   e. Eliminar un estudiante de la base de datos.

2. Utiliza listas de Python para almacenar la información de los estudiantes. Por ejemplo, puedes crear una lista de diccionarios, donde cada diccionario representa un estudiante con sus datos.

3. Ofrece un menú interactivo que permita a los usuarios elegir las acciones que desean realizar en la base de datos.

4. Implementa la lógica necesaria para cada acción, como agregar, buscar, actualizar o eliminar estudiantes.

## Ejemplo de Base de Datos

A continuación, se muestra un ejemplo de cómo podría verse una base de datos de estudiantes utilizando listas y diccionarios:

```python
base_de_datos = [
    {"nombre": "Juan", "id": 1, "edad": 20},
    {"nombre": "María", "id": 2, "edad": 22},
    {"nombre": "Carlos", "id": 3, "edad": 21}
]

Desarrolle el código en la siguiente lista.

````python
# Lista donde se almacenará los estudiantes
base_de_datos = []

# Id que va ir autoincrementandose en cada registro ingresado
id_estudiante = 0

# Método para realizar una busqueda secuencial de un estudiante
def buscar_estudiante(id_estudiante):
  for estudiante in base_de_datos:
    if estudiante["id"] == id_estudiante:
      return estudiante
  return None

# Método para listar los estudiantes
def listar_estudiantes():
  if not base_de_datos:
    print("\nNo hay estudiantes ingresados.")
  else:
    print("\nListado de estudiantes:")
    for estudiante in base_de_datos:
      print(f"ID: {estudiante['id']}, Nombre: {estudiante['nombre']}, Edad: {estudiante['edad']}")
      print("--------------------------------------")

# Método para crear un estudiante
def crear_estudiante():
  global id_estudiante
  try:
    nombre = input("Ingrese el nombre del estudiante: ")
    edad = int(input("Ingrese la edad del estudiante: "))
  except ValueError:
    print("\nIngrese correctamente los datos.")
    return
  id_estudiante += 1
  estudiante = {"nombre": nombre, "id": id_estudiante, "edad": edad}
  base_de_datos.append(estudiante)
  print("\nEstudiante ingresado exitosamente!")

# Método para actualizar un estudiante
def actualizar_estudiante():
  try:
    id_estudiante = int(input("Ingrese el ID del estudiante a actualizar:"))
  except ValueError:
    print("\nEl ID debe ser un número entero.")
    return

  estudiante = buscar_estudiante(id_estudiante)
  if estudiante:
    nombre = input("Ingrese el nuevo nombre del estudiante: ")
    edad = int(input("Ingrese la nueva edad del estudiante: "))
    estudiante["nombre"] = nombre
    estudiante["edad"] = edad
    print("\nEstudiante actualizado exitosamente!")
  else:
    print("\nEstudiante no encontrado.")

# Método para eliminar un estudiante de la base de datos
def eliminar_estudiante():
  try:
    id_estudiante = int(input("Ingrese el ID del estudiante a eliminar: "))
  except ValueError:
      print("\nEl ID debe ser un número entero.")
      return

  # Buscar y eliminar el estudiante de la base de datos
  estudiante = buscar_estudiante(id_estudiante)
  if estudiante:
    base_de_datos.remove(estudiante)
    print("\nEstudiante eliminado exitosamente!")
  else:
    print("\nEstudiante no encontrado.")

# Menú principal
def menu():
  while True:
     print("\n============== Menú Estudiantes ======================")
     print("1. Agregar un nuevo estudiante")
     print("2. Mostrar la lista de estudiantes")
     print("3. Buscar un estudiante por su número de identificación")
     print("4. Actualizar la información de un estudiante existente")
     print("5. Eliminar un estudiante de la base de datos")
     print("6. Salir")

     opcion = input("Seleccione una opción: ")
     match opcion:
        case "1":
          crear_estudiante()
        case "2":
          listar_estudiantes()
        case "3":
          try:
            id_estudiante = int(input("Ingrese el ID del estudiante a buscar: "))
          except ValueError:
            print("\nEl ID debe ser un número entero.")
            continue
          estudiante = buscar_estudiante(id_estudiante)
          if estudiante:
            print("\nEstudiante econtrado exitosamente!")
            print("Información del estudiante:")
            print(f"ID: {estudiante['id']}, Nombre: {estudiante['nombre']}, Edad: {estudiante['edad']}")
            print("-------------------------------------------")
          else:
            print("\nEstudiante no encontrado.")
        case "4":
          actualizar_estudiante()
        case "5":
          eliminar_estudiante()
        case "6":
          print("\nSaliendo del programa...")
          break
        case _:
          print("\nOpción inválida. Por favor, seleccione una opción válida.")

menu()
````

# Actividad: Rebanado de listas.

Revisa la unidad 10,sección 5.3 aparece un resumen de los comandos para tomar secciones de una lista, aplica los comandos en la siguiente lista:


list_notas = [15, 21, 24, 40, 13, 14, 18, 17, 14.5]



````python
# Espacio para desarrollar las pruebas, si tienes alguna duda o comentario adicional
# lo puedes dejar planteado acá.

list_notas = [15, 21, 24, 40, 13, 14, 18, 17, 14.5]

# Rebanar los elementos desde el inicio hasta(fin-1)(no incluye los elementos del índice final)
# list_notas[inicio:fin]
print(list_notas[1:5])

# Rebanar desde el principio hasta el final de la lista. Rebanar la parte final de la lista
# list_notas[start:]
print(list_notas[1:])

# Rebanar desde el inicio hasta el final-1 índice
# list_notas[:end]
print(list_notas[:-1])

# Rebanar toda la lista
# list_notas[:]
print(list_notas[:])

# Rebanar saltándose un paso de principio a fin-1
# list_notas[inicio:fin:paso]
print(list_notas[::2])

# Rebanar dos elementos del extremo
# list_notas[-2:]
print(list_notas[-2:])

# Rebanar partidas enteras desde el principio excepto las dos últimas partidas
# list_notas[:-2]
print(list_notas[:-2])

# Importar todos los artículos pero cortando en orden inverso
# list_notas[::-1]
print(list_notas[::-1])

# Cortar los dos primeros elementos en orden inverso
# list_notas[1::-1]
print(list_notas[1::-1])
````


# Actividad: Diferentes métodos usados en listas.

De la unidad 10, sección 6.3 revise los métodos disponibles para usar sobre las listas.

Aplique los métodos disponibles sobre la lista mostrada en la siguiente celda.

Nota: usando el método **sort()** ordene de mayor a menor y viceversa (Si es necesario revise documentación adicional relacionada sobre este método)


````python
# Aplique los métodos disponibles y agregue comentarios breves sobre su funcionamiento.

list_notas = [15, 21, 24, 40, 13, 14, 18, 13, 17, 14.5]


# Espacio para desarrollo del código y comentarios.

# Busca un lugar utilizando el elemento x.
# index(x)
# Devuelve el índice del elemento(x) buscado
print(list_notas.index(13))

# Añade el elemento x al final de la lista
# append(x)
list_notas.append(20)
print(list_notas)
# Agrega otra lista como un único elemento
list_notas.append([22, 23])
print(list_notas)

# Devuelve la cantidad de veces que aparece el elemento (x) en la lista
# count(x)
elemento = 13
conteo = list_notas.count(elemento)
print(f"El número {elemento} aparece {conteo} veces en la lista.")

# Añadir multiples elementos de un iterable (lista, tupla o cadena) al final de una lista existente.
# extend[x1, x2]
list_notas.extend([22, 23])
print(list_notas)

# Añadir un elemento(x) indicando la posición deseada.
# insert(index, x)
list_notas.insert(2, 100)
print(list_notas)

# Eliminar un elemento (x) de la lista
# remove(x)
list_notas.remove(100)
print(list_notas)

# Eliminar un elemento dada una posición y luego devolverlo
# Si se omite el índice, borrará y devolverá el último de la lista
# pop(index)
elemento_eliminado = list_notas.pop(-3)
print(f"Elemento eliminado: {elemento_eliminado}")
print(list_notas)

# Ordenar los valores en orden ascendente
list_notas.sort()
print("Lista ordenada ascendente: ", list_notas)
list_notas.sort(reverse=True)
print("Lista ordenada descendente: ", list_notas)

# Invertir los elementos de la lista
# reverse()
list_notas.reverse()
print("Lista invertida: ", list_notas)
````

# Actividad: Crear diccionarios y guardarlos como archivo JSON.

De la unidad 11 sección 4.6 revise el ejemplo mostrado.

Genera un diccionario, sobre la información que deseas guardar como base de datos, desarrollo el código necesario para generar el archivo JSON, y también el código para cargar la información del archivo JSON.


````python
# Espacio para desarrollar el código.

import json

# Diccionario de la data
data = {
    "estudiantes":[
        {"id": 1, "nombre": "Paul", "edad": 21},
        {"id": 2, "nombre": "Josue", "edad": 21},
        {"id": 3, "nombre": "Johnny", "edad": 21}
    ]
}


# Guardar el diccionario en un archivo JSON
with open('estudiantes.json', 'w') as f:
  # (ensure_ascii = False) mantiene los caracteres especiales tal cual en el JSON
  json.dump(data, f, indent='\t', ensure_ascii=False)
print("Archivo JSON creado exitosamente!")

# Leer la data desde el archivo JSON y mostrar su contenido
with open("estudiantes.json", "r") as f:
  data_cargada = json.load(f)

print("\nContenido del archivo JSON:")
print(json.dumps(data_cargada, indent='\t', ensure_ascii=False))
````

# Actividad: Revisión y manejo de listas bidimensionales.

## Instrucciones

1. Describa brevemente que entiende por lista bidimensional.
2. Crear una lista bidimensional.
3. Generar un doble loop para acceder de la lista bidimensional
4. Sumar todo los elementos de esta lista bidimensional.

````python
# Espacio para desarrollar el código, escriba comentarias breves para explicar su funcionamiento.

# 1.
# Una lista bidimensional es una estructura de datos que organiza los elementos en filas y columnas.
# Una lista bidimensional es una lista que contiene otras listas en su interior,
# donde cada lista interna representa una fila y cada elemento dentro de esa lista representa una columna.

# 2. Creando una lista bidimensional
lista_bidimensional = [ [2, 4, 6, 8], [1, 3, 5, 7] ]


# 3. Recorrer con doble loop anidado
def loop_anidado():
  for fila in lista_bidimensional:
    for elemento in fila:
      print(elemento, end=" ")
    print()

loop_anidado()

# 4. Sumar los elementos de la lista bidimensional
def sumar_elementos():
  suma = 0
  for fila in lista_bidimensional:
    for elemento in fila:
      suma += elemento
  print(f"La suma de los elementos es: {suma}")

sumar_elementos()
````

# Actividad: Revision de los métodos usados en el manejo de diccionarios.

## Instrucciones

1. Crear un diccionario.
2. Testear los siguientes métodos para el majeo de diccionarios get(), keys(), values(), items(), pop(), clear()
3. Agregar un breve comentario sobre su funcionamiento.

````python
# Espacio para desarrollar el código.

estudiante = {
    "id": 1,
    "nombre": "Paul",
    "edad": 21,
    "carrera": "Ingeniería en Tecnologías de la Información"
}

# Método para obtener un valor de una clave dada
# get(clave, valor_predeterminado)
# valor_predeterminado (opcional)
print(estudiante.get("nombre"))
print(estudiante.get("sexo", "No encontrado"))

# Método para obtener todas las claves del diccionario
# keys()
print(estudiante.keys())

# Método para obtener todos los valores del diccionario
# values()
print(estudiante.values())

# Método para obtener los pares (clave, valor)
# items()
print(estudiante.items())

# Método para eliminar y devolver un valor asociado a una clave
# pop(clave, valor_predeterminado)
# valor_predeterminado (opcional)
print("Valor eliminado: ", estudiante.pop("carrera"))
print(estudiante.pop("sexo", "No encontrado"))

# Método para eliminar todos los elementos (pares clave-valor)
# clear()
estudiante.clear()
print(estudiante)
````
