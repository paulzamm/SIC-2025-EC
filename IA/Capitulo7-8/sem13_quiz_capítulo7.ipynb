### ¿Cuál de las siguientes afirmaciones sobre la minería de textos es incorrecta?
1. Puedes conocer la reacción ante un grupo específico.

## Q 07-05

### Explique sobre TF (Frecuencia de términos) y escriba cómo calcular TF (fórmula).

**TF** es una métrica numérica que indica con qué frecuencua aparece una palabra específica dentro de un documento concreto.

- Fórmula para calcular TF:
$$
TF(t, d) = \frac{\text{Frecuencia del término } t \text{ en el documento } d}{\text{Número total de términos en el documento } d}
$$

t = el término o palabra que estamo analizando.

d = el documento específico.

- Ejemplo:

Frase = "Amo aprender python y usar python para datos."

1. **Término (t)**: "Python"
2. **Conteo de aparición**: Aparece 2 veces.
3. **Total de palabras (d)**: La frase tiene 7 palabras en total.

Cálculo:
$$
TF("Python", frase) = \frac{2} {7} = 0.285
$$
