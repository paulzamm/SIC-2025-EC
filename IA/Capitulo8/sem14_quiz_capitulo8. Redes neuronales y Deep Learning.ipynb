## Q 08-03

### ¿Qué sucede si el error de verificación aumenta constantemente cuando graficamos el error de verificación para cada época utilizando el método de descenso de pendiente por lotes? Además, ¿cómo se puede solucionar este problema?

Esto se conoce como **Sobreajuste**. Esto ocurre cuando la red neuronal aorende demasiado bien los datos de entrenamiento, llegando a memorizar el ruido y los detalles específicos en lugar de los patrones generales.

- Solución: **Parada Temprana**, consiste en detener el proceso de entrenamiento justo en el punto de inflexión donde el error de verificación deja de disminuir y comienza a subir.

## Q 08-04

### La siguiente figura son datos con una estructura tridimensional.
1) Presente las ecuaciones de convolución para (a)
2) Presente las ecuaciones de convolución para (b)

image.png

1. Ecuación de convolución para (1): Datos Multicanal (Convolución 2D): $$y_{i,j} = \sum_{c=0}^{k-1} \sum_{u=0}^{h-1} \sum_{v=0}^{h-1} x_{(i+u), (j+v), c} \cdot w_{u, v, c} + b$$

Donde:
- $y_{i,j}$: Valor de salida en la posición $(i, j)$.
- $k$: Número total de canales (profundidad de entrada).
- $h$: Tamaño espacial del kernel (altura y anchura).
- $x$: Valor de entrada.$w$: Peso del kernel.
- $b$: Sesgo (bias).
- **Nota: Se suman los productos de todos los canales $c$**.

2. Ecuación de convolución para (b): Datos 3D (Convolución 3D): $$y_{z,i,j} = \sum_{d=0}^{D-1} \sum_{u=0}^{h-1} \sum_{v=0}^{h-1} x_{(z+d), (i+u), (j+v)} \cdot w_{d, u, v} + b$$

Donde: 

- $y_{z,i,j}$: Valor de salida en la posición tridimensional $(z, i, j)$.
- $D$: Profundidad del kernel (en la imagen se etiqueta como parte de la dimensión del kernel).
- $h$: Tamaño espacial del kernel.
- $z$: Índice de la posición en el eje de profundidad/tiempo.
- **Nota: A diferencia de (a), aquí el kernel se mueve a través del eje $z$, generando un volumen de salida.**

## Q 08-05

### Aplique la función softmax cuando la salida de la red neuronal sea $(0.4,2.0,0.001,0.32)^T$ y escriba el resultado.

La fórmula de Softmax para cada elemento $z_i$ es:
$$\sigma(z)_i = \frac{e^{z_i}}{\sum_{j=1}^K e^{z_j}}$$

```python
import numpy as np
from scipy.special import softmax

z = np.array([0.4, 2.0, 0.001, 0.32])
result = softmax(z)
print(np.round(result, 4))
```

La red neuronal está un **65.6%** segura de que la respuesta correcta es la segunda clase con un valor de 2.0
