## Q 06-04

### Escribe tres o más ejemplos de algoritmo de clustering.

- **K-Means**: Es el algoritmo de particionamiento más famoso. Divide los daots en k grupos predefinidos, asignando cada punto al **centroide** más cercano. Funciona bien con grupos esféricos y separados.
- **DBSCAN**: Algoritmo basado en **densidad**. Agrupa puntos que están muy juntos y marca como **ruido** a los puntos que están en zonas de baja densidad. Es capaz de encontrar clusters con formas irregulares.
- **Agrupamiento Jerárquico**: Crea una estructura de árbol de los grupos.
    - Aglomerativo: Empieza considerando cada punto como un grupo individual y los va fusioanando hasta formar uno solo.
    - Divisivo: Empieza con todos los puntos en un solo grupo y los va dividiendo.
- **Modelos de Mezcla Gaussiana**: Es una versión probabilistica de K-Means. Asume que los datos se generan a partir de una mezcla de varias distribuciones gaussianas. Permite que un punto perteneza (Un 60% al grupo A y un 40% al grupo B).

## Q 06-05

### Explique los dos métodos para elegir el número óptimo de cluters cuando se utilizan k-medias. Describa cómo encontrar el número óptimo de clusters utilizando los dos métodos.

1. **El método del Codo**

Es el método más intuitivo y visual. Se basa en medir la Inercia (Suma de errores cuadráticos), que es la distancia total de los puntos hacia el centro de su cluster.

- Funcionamiento:
    - Ejecuta K-Means varias veces, variando k
    - Calcula la inercia para cada k. Naturalmente, a medida que aumenta k, el error baja.
    - Gráfica resultados: Eje X = Numero de cluster (k), Eje Y = Inercia.
- Cómo encontrrar el número óptimo: Buscar el punto donde la curva deja de descender bruscamente y comienza a aplanarse. Ese punto de inflexión se asemeja a un **codo** de un brazo flexionado. EL valor de k en ese codo es el óptimo, ya que agregar más clusters después de ese punto no aporta una mejora significativa en la calidad del agrupamiento, solo aumenta la complejidad.

2. **El método de la Silueta**

Este método es más rigurosos matemáticamente y menos subjetivo que mirar un gráfico a ojo. Mide la calidad del agrupamiento calculando qué tan parecido es un punto a su propio grupo comparando con qué tan parecido es a los grupos vecinos.

El coeficiento de silueta va de -1 a 1:
- **+1**: El punto está perfectamente clasificado
- **0**: El punto está en el borde entre dos grupos
- **-1**: El punto está en el grupo 

Conclusión:
- Usar el codo para una estimación rápida y visual.
- Usar la siluetaa cuando el gráfico del codo sea ambiguo o validar métrica más precisa.


## Q 06-06

### En un conjunto de datos bidimensional con tres grupos de formas elípticas, ¿qué criterio práctico ayuda más a decidir entre clustering jerárquico y no jerárquico?

1)Si los clusters tienen tamaños y densidades muy diferentes, preferir jerárquico.

2)Si los clusters son de forma no esférica, preferir no jerárquico (p. ej., K-means).

3)Si se espera una estructura de agrupamiento basada en densidad, preferir jerárquico.

4)Si la forma de los clusters es elíptica y la densidad varía, probablemente sea mejor un método no jerárquico o basado en densidad.

La respuesta correcta es la **1**
- Si los grupos tienen formas, tamaños o densidades irregulares, K-Means fallará, por lo que es preferible optar por un método jerárquico.
